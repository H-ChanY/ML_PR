# 데이터 슬라이싱

<p>DataFrame 바로 뒤의 [] 연산자는 넘파이의 []나 Series의 []와는 다르다.
<p>DataFrame 바로 뒤의 [] 내 입력 값은 칼럼명(또는 칼럼의 리스트)을 지정해 
칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용해야함.
<p>DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터를 추출하는 방법은 사용하지 않는 게 좋음.


```python
# loc[] 연산자 명칭 기반으로 데이터를 추출함. 따라서 행 위치에는 DataFrame index값을 그리고 열 위치에는 칼럼 명을 입력해준다.\
laptops_df
```

```python
data={
    'Name':['heo','chan','yong'],
    'Year':[19,98,3],
    'Gender':['M',"M",'M']
}
data_df=pd.DataFrame(data,index=[0,1,2])
data_df
```

```python
data_df.ix[0,0]
```

```python
data_df.iloc[0,2] # 첫번째는 행 두번째는 열 loc처럼 문자열이나 그런거 안됨.
```

## 책을 보다 보면 ix라는 연산이 있는데 이거는 업데이트 데면서 loc나 iloc로 대체 되었음 그래서 작동이 안되는거임.
