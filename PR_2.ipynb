### 영수증

```python
import random
상품=['Apple','Banana','Peach']
가격=[3000,2000,1000]
```

```python
i=0
for v in range(50):
    sale_list=[]
    num=random.randint(0,2)
    i+=1
    sale_list.append(i)
    sale_list.append(상품[num])
    sale_list.append(가격[num])
    print(sale_list)
```
